```python

def get_labels(row):
    labels = []
    max_val = -1
    max_label = ''
    label_dict = {}
    for col in total_cols:
        if row[col] > max_val:
            max_val = row[col]
            max_label = col.split('_')[-1] 

        if row[col] > 0.5:
            labels.append(col.split('_')[-1])
            label_dict[col.split('_')[-1]] = row[col]
    
    if len(labels) == 0:
        labels.append(max_label)
        
    if len(labels) > 3:
        sorted_dict = dict(sorted(label_dict.items(), key=lambda item: item[1], reverse=True))
        print(sorted_dict)
        labels = []
        labels.extend(list(sorted_dict.keys())[0:3])
        
    return str(labels)
```

```python
test['Product_Holding_B2'] = pred.apply(get_labels, axis=1 )
```

```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('train_updated.csv')
test = pd.read_csv('test_updated.csv')
```

```python
B1_cols = [col for col in list(train.columns) if col.startswith('B1')]
B2_cols = [col for col in list(train.columns) if col.startswith('B2')]
```

```python
B1_cols
```

```python
B2_cols
```

```python
from sklearn.model_selection import train_test_split

target_cols = B2_cols
X = train.drop(columns = target_cols)
y = train[target_cols]

```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
X_train
```

```python
from sklearn.multioutput import MultiOutputClassifier
import catboost
classifier = MultiOutputClassifier(catboost.CatBoostClassifier(verbose=1))
```

```python

classifier.fit(X_train, y_train)
```

```python
print(classifier.score(X_test, y_test))
```

```python
preds = classifier.predict_proba(test)
preds
```

```python
preds_cal = []
for item in preds:
    preds_cal.append( [local_item[1] for local_item in item] )
```

```python
df_catboost = pd.DataFrame(np.array(preds_cal).T, columns=B2_cols)
df_catboost.to_csv('catboost_preds.csv', index=False)
```
